```python
import math
```

```python


# Configuração do Tabuleiro
tabuleiro = [' ' for _ in range(9)]

def verificar_vitoria(b, p):
    vitorias = [[0,1,2], [3,4,5], [6,7,8], [0,3,6], [1,4,7], [2,5,8], [0,4,8], [2,4,6]]
    return any(all(b[i] == p for i in combo) for combo in vitorias)

def tabuleiro_cheio(b):
    return ' ' not in b

# O Algoritmo Minimax
def minimax(b, profundidade, is_maximizing):
    if verificar_vitoria(b, 'O'): return 1
    if verificar_vitoria(b, 'X'): return -1
    if tabuleiro_cheio(b): return 0

    if is_maximizing:
        melhor_score = -math.inf
        for i in range(9):
            if b[i] == ' ':
                b[i] = 'O'
                score = minimax(b, profundidade + 1, False)
                b[i] = ' '
                melhor_score = max(score, melhor_score)
        return melhor_score
    else:
        melhor_score = math.inf
        for i in range(9):
            if b[i] == ' ':
                b[i] = 'X'
                score = minimax(b, profundidade + 1, True)
                b[i] = ' '
                melhor_score = min(score, melhor_score)
        return melhor_score

def melhor_jogada():
    melhor_score = -math.inf
    jogada = -1
    for i in range(9):
        if tabuleiro[i] == ' ':
            tabuleiro[i] = 'O'
            score = minimax(tabuleiro, 0, False)
            tabuleiro[i] = ' '
            if score > melhor_score:
                melhor_score = score
                jogada = i
    return jogada

# Loop Simples de Jogo
print("Você é o 'X'. Escolha de 0 a 8:")
while not tabuleiro_cheio(tabuleiro):
    # Jogada do Humano
    jogada_h = int(input("Sua vez (0-8): "))
    if tabuleiro[jogada_h] != ' ': continue
    tabuleiro[jogada_h] = 'X'

    if verificar_vitoria(tabuleiro, 'X'):
        print("Você venceu!"); break

    if tabuleiro_cheio(tabuleiro): break

    # Jogada da IA
    print("IA pensando...")
    tabuleiro[melhor_jogada()] = 'O'

    # Exibir Tabuleiro
    for i in range(0, 9, 3):
        print(f"| {tabuleiro[i]} | {tabuleiro[i+1]} | {tabuleiro[i+2]} |")

    if verificar_vitoria(tabuleiro, 'O'):
        print("IA venceu!"); break
```
